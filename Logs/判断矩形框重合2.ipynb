```python
import pandas as pd
cv=pd.read_csv('./cv_info.log',header=None)
```

```python
cv['time']=cv.iloc[:,0].apply(lambda x : x.split('|')[0].split('.')[0].strip()).astype('datetime64[ns]')#pd.to_datetime(,format='%Y-%m-%d %H:%M:%s')
cv['cv_x1']=cv.iloc[:,0].apply(lambda x : float(x.split('=')[-1].strip()))
cv['cv_y1']=cv.iloc[:,1].apply(lambda x : float(x.split('=')[-1].strip()))
cv['cv_x2']=cv.iloc[:,2].apply(lambda x : float(x.split('=')[-1].strip()))
cv['cv_y2']=cv.iloc[:,3].apply(lambda x : float(x.split('=')[-1].strip()))
cv_single=cv.drop_duplicates(['time'],keep ='first')
cv_single=cv_single[['time','cv_x1','cv_y1','cv_x2','cv_y2']].reset_index(drop=True)
cv_single.to_csv('cv.csv')
```

```python
yolo=pd.read_csv('./yolo_info.log',header=None)
yolo['time']=yolo.iloc[:,0].apply(lambda x : x.split('|')[0].split('.')[0].strip()).astype('datetime64[ns]')#pd.to_datetime(,format='%Y-%m-%d %H:%M:%s')
yolo['yolo_x1']=yolo.iloc[:,0].apply(lambda x : float(x.split('=')[-1].strip()))
yolo['yolo_y1']=yolo.iloc[:,1].apply(lambda x : float(x.split('=')[-1].strip()))
yolo['yolo_x2']=yolo.iloc[:,2].apply(lambda x : float(x.split('=')[-1].strip()))
yolo['yolo_y2']=yolo.iloc[:,3].apply(lambda x : float(x.split('=')[-1].strip()))
yolo_single=yolo.drop_duplicates(['time'],keep ='first')
yolo_single=yolo_single[['time','yolo_x1','yolo_y1','yolo_x2','yolo_y2']].reset_index(drop=True)
yolo_single.to_csv('yolo.csv')
```

```python
yolo_s=yolo_single.set_index('time')
cv_s=cv_single.set_index('time')
data=pd.concat([yolo_s,cv_s],axis=1,join='inner')
```

```python
# 重合矩形判断，左下右上两个点，左下为原点坐标
# def solve(R1, R2):
#     if (R1[0]>=R2[2]) or (R1[2]<=R2[0]) or (R1[3]<=R2[1]) or (R1[1]>=R2[3]):
#         return False
#     else:
#         return True

# 重合矩形判断，左上右下两个点，左下为原点坐标
def iou(R1,R2):
    x1=max(R1[0],R2[0])
    y1=max(R1[1],R2[1])
    x2=min(R1[2],R2[2])
    y2=min(R1[3],R2[3])
    inter=abs(x1-x2)*abs(y1-y2)
    if inter>0:
        x1=min(x1,x2)
        x2=max(x1,x2)
        y1=max(y1,y2)
        y2=min(y1,y2)
        return True,[x1,y1,x2,y2]
    else:
        return False,[]
```

```python
# 结果
res=[]
for i in range(data.shape[0]):
    tem=iou(data.iloc[i,0:4].tolist(),data.iloc[i,4:].tolist())
    if tem[0]:
        res.append(str(data.index[i])+'有重合❌'+'，重合坐标为'+str(tem[1]).replace('[','').replace(']',''))
    else:
        res.append(str(data.index[i])+'无重合✅')
res
```
